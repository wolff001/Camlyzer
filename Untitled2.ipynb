```python
from google.colab import files
import pytesseract
import wikipedia
from PIL import Image
import torch
from typing import TypedDict
from transformers import (
    AutoTokenizer, AutoModelForCausalLM,
    BlipProcessor, BlipForConditionalGeneration
)
from peft import PeftModel
from langgraph.graph import StateGraph, END
from google.colab import files
```

```python
# ✅ 1. Define Typed State for LangGraph
class ChatState(TypedDict):
    image: Image.Image
    caption: str
    ocr: str
    web: str
    question: str
    answer: str

# ✅ 2. Load Caption Model (BLIP)
blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to("cuda")

# ✅ 3. Load Mistral + QLoRA Model
model_name = "mistralai/Mistral-7B-Instruct-v0.1"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token

base_model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    load_in_4bit=True,
    trust_remote_code=True
)

```

```python
model = PeftModel.from_pretrained(base_model, "/content/qlora-chatbot-model/qlora-chatbot-model")
model.eval()
```

```python
def caption_node(state: ChatState) -> ChatState:
    inputs = blip_processor(state["image"], return_tensors="pt").to("cuda")
    output = blip_model.generate(**inputs)
    state["caption"] = blip_processor.decode(output[0], skip_special_tokens=True)
    return state



import pytesseract
from PIL import Image

def ocr_node(state: ChatState) -> ChatState:
    # Step 1: Convert to grayscale (for better OCR)
    image = state["image"].convert("L")

    # Step 2: Run OCR
    raw_text = pytesseract.image_to_string(image)

    # Step 3: Clean text (keep only readable lines)
    lines = [line.strip() for line in raw_text.split("\n") if len(line.strip()) >= 3]
    filtered = "\n".join(lines)

    # Step 4: If no clean lines, skip it
    if not filtered or len(filtered.split()) < 3:
        state["ocr"] = ""  # Leave it empty like you said
    else:
        state["ocr"] = filtered

    return state





def web_node(state: ChatState) -> ChatState:
    try:
        caption_words = list(dict.fromkeys(state["caption"].split()))
        caption_keywords = [w for w in caption_words if len(w) > 2][:3]
        query = " ".join(caption_keywords) if caption_keywords else "safety"

        if len(set(caption_keywords)) <= 1:
            raise ValueError("Caption too weak")

        summary = wikipedia.summary(query, sentences=2)
        state["web"] = summary

    except Exception:
        state["web"] = "No relevant web info found."
    return state




def qa_node(state: ChatState) -> ChatState:
    prompt = f"""You are an assistant that answers questions about what's in an image.

Image Caption: {state['caption']}
Text in Image (OCR): {state['ocr'] or 'No text found.'}
Relevant Info: {state['web'] or 'None'}

Now answer this clearly and briefly based on the actual content in the image:
Question: {state['question']}

Answer:"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=200,
            temperature=0.7,
            top_p=0.9,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    if "Answer:" in response:
        state["answer"] = response.split("Answer:")[-1].strip()
    else:
        state["answer"] = response.strip()

    return state






```

```python
graph = StateGraph(ChatState)
graph.add_node("caption", caption_node)
graph.add_node("ocr", ocr_node)
graph.add_node("web", web_node)
graph.add_node("qa", qa_node)

graph.set_entry_point("caption")
graph.add_edge("caption", "ocr")
graph.add_edge("ocr", "web")
graph.add_edge("web", "qa")
graph.add_edge("qa", END)

app = graph.compile()
```

```python
while True:
    print("\n📤 Upload an image (or type 'exit' to quit):")
    uploaded = files.upload()
    if not uploaded:
        break

    image_path = list(uploaded.keys())[0]
    if "exit" in image_path.lower():
        break

    image = Image.open(image_path).convert("RGB")

    # Step 1: Process image once (caption + ocr + web)
    state: ChatState = {
        "image": image,
        "caption": "",
        "ocr": "",
        "web": "",
        "question": "",
        "answer": ""
    }

    # Run LangGraph up to web
    temp_graph = StateGraph(ChatState)
    temp_graph.add_node("caption", caption_node)
    temp_graph.add_node("ocr", ocr_node)
    temp_graph.add_node("web", web_node)
    temp_graph.set_entry_point("caption")
    temp_graph.add_edge("caption", "ocr")
    temp_graph.add_edge("ocr", "web")
    temp_graph.add_edge("web", END)
    preprocess = temp_graph.compile()

    # Invoke image → caption, ocr, web
    state = preprocess.invoke(state)

    # Step 2: Print all state values
    print("\n📊 STATE CONTENT:")
    print("🖼️ Caption:", state["caption"])
    print("🔠 OCR:", state["ocr"])
    print("🌐 Web Info:", state["web"])

    # Step 3: Ask multiple questions on same image
    while True:
        question = input("\n❓ Ask a question about this image (or 'next' to upload another): ").strip()
        if question.lower() in ["exit", "next"]:
            break

        state["question"] = question
        state["answer"] = ""

        final_state = qa_node(state)

        print("\n🤖 Assistant:", final_state["answer"])
```
